```
from time import time
def ejemplo1( n ):
    start_time = time()
    c = n + 1
    d = c * n
    e = n * n
    total = c + e - d
    #print(f"total={ total }")
    elapsed_time = time() - start_time
    print(" %0.10f s" % elapsed_time)
for n in  range (100,1100,100 ):
  ejemplo1( n )

  #T(n) = 4 
```

```
from time import time
def ejemplo2( n ):
    start_time = time()
    contador = 0
    for i in range( n ) :
        for j in range( n ) :
            contador += 1
    elapsed_time = time() - start_time
    print(" %0.10f s" % elapsed_time)
    return contador

for n in  range (100,1100,100 ):
  ejemplo2( n )

  #T(n) = 1+n^2
```

```
from time import time
def ejemplo4( n ):
    start_time = time()
    x = 3 * 3.1416 + n
    y = x + 3 * 3 - n
    z = x + y
    elapsed_time = time() - start_time
    print(" %0.10f s" % elapsed_time)
    return z

for n in  range (100,1100,100 ):
  ejemplo4( n )

  # 3
```

```
from time import time
def ejemplo5( x ):
    start_time = time()
    n = 10
    for j in range( 0 , x , 1 ): # En un rango de 0 a X y que va de 1 en 1
        n = j + n
    elapsed_time = time() - start_time
    print(" %0.10f s" % elapsed_time)
    return n

for n in  range (100,1100,100 ):
  ejemplo5( n )

  #1+x
```
